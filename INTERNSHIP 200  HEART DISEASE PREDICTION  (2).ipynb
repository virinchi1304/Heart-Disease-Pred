```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("heart.csv")
dataset
```

```python
x=dataset.drop('target',axis='columns')
y=dataset['target']
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=5)
```

```python
y_train
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
log_reg=LogisticRegression(random_state=0,max_iter=900)
```

```python
log_reg = LogisticRegression(dual=False ,solver='newton-cg')
```

```python
log_reg.fit(x_train,y_train)
```

```python
y_pred=log_reg.predict(x_test)
y_pred
```

```python
print(classification_report(y_pred,y_test))
```

```python
# K NN MODEL
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 5).fit(x_train,y_train)
knn.score(x_test,y_test)
```

```python
error_rates=[]
```

```python
for k in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    pred=knn.predict(x_test)
    error_rates.append(np.sum(y_test!= pred))
```

```python
plt.plot(list(range(1,40)),error_rates,marker='o',ls ='--',mfc='red')
```

```python
k=error_rates.index(np.min(error_rates))+1
k
```

```python
knn=KNeighborsClassifier(n_neighbors= 10,algorithm='auto',leaf_size=3 ,p=1,metric='manhattan')
knn.fit(x_train, y_train)
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from sklearn.ensemble import RandomForestClassifier
  
#importing datasets  
dataset= pd.read_csv('heart.csv')  
  
#Extracting Independent and dependent Variable  
x=dataset.drop('target',axis='columns')
y=dataset['target']
  
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0)  
  
```

```python

forest = RandomForestClassifier(n_estimators=109)
forest.fit(x_train, y_train)
```

```python
y_pred = forest.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test, y_pred)
```
